```python
from zlib import crc32
import numpy as np
def check(unique_id,test_ratio):
    return crc32(np.int64(unique_id)) & 0xffffffff < test_ratio*2**32
def split_train_test(data,test_ratio,unique_col):
    ids = data[unique_col]
    test_set_id = ids.apply( lambda id_ : check(id_,test_ratio))
    return data.loc[test_set_id] , data.loc[~test_set_id]
student_unq_col = student.reset_index()
test_set,train_set = split_train_test(student_unq_col,0.2,"index")
```

```python
import sklearn
from sklearn.preprocessing import OneHotEncoder,StandardScaler 
from sklearn.compose import ColumnTransformer
categorial_encoder = OneHotEncoder()
x_test = x_test.drop("index",axis=1,errors="ignore")
cat_attributes = [ "school" , "sex" ,"address" , "famsize","Pstatus","Mjob","Fjob","reason","guardian","schoolsup",
                  "famsup","paid","activities" ,"nursery","higher","internet","romantic" ]
num_attributes = [c for c in x_train.columns if c not in cat_attributes]
pipeline1 = ColumnTransformer([
                ("num",StandardScaler(),num_attributes),
                ("cat",OneHotEncoder(),cat_attributes)])
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
pipeline = Pipeline([ ("preprocessing",pipeline1),
                        
    ( "model" , LinearRegression()),
                   
                        
                    ])

pipeline.fit(x_train,y_train)
predictions = pipeline.predict(x_test)
lin_mse = mean_squared_error(y_test,predictions)
rmse = np.sqrt(lin_mse)
print(rmse)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
pipeline = Pipeline([ ("preprocessing",pipeline1),
                        
    ( "model" , RandomForestRegressor()),])

param = [{
    'model__bootstrap': [True],
    'model__n_estimators': [100, 200],
    'model__max_features': [2,4,6,8],
    'model__max_depth': [5,10,20],
    'model__min_samples_leaf': [1,2,5],
    'model__min_samples_split': [2,5,10]
}]
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(pipeline,param,cv=5,scoring='neg_mean_squared_error',return_train_score = True,n_jobs= -1 ,verbose = 2)
grid_search.fit(x_train,y_train)
best_model = grid_search.best_estimator_
prediction = best_model.predict(x_test)
forest_rmse = mean_squared_error(y_test,prediction)
rmse = np.sqrt(forest_rmse)
print(rmse)
```

```python
predictions = grid_search.predict(x_train)
train_rmse = mean_squared_error(y_train,predictions)
rmse1 = np.sqrt(train_rmse)
print(rmse1)
```
